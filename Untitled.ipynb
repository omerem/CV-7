```python
from numpy import mgrid
import math
import numpy as np
def bilinear_interpolation (f):
    def interpolate (y,x):
        x_ceil = math.ceil(x)
        x_floor = math.floor(x)
        y_ceil = math.ceil(y)
        y_floor = math.floor(y)
        
        x_frac = x-math.floor(x)
        y_frac = y-math.floor(y)
        
        if(x_ceil > 79 or x_floor < 0 or y_ceil > 79 or y_floor < 0):
            return 0
        return (f(x_floor, y_floor) * (1-x_frac) * (1-y_frac)
               + f(x_ceil, y_floor) * (x_frac) * (1-y_frac)
               + f(x_floor, y_ceil) * (1-x_frac) * (y_frac)
               + f(x_ceil, y_ceil) * (x_frac) * (y_frac))
    return interpolate
```

```python
from matplotlib import pyplot as plt
plt.show()
```

```python
def compute_T_inverse(v,u):
    def T_inverse (y_tilde, x_tilde):
        return np.array((y_tilde - v[y_tilde, x_tilde], x_tilde-u[y_tilde, x_tilde]))
    return T_inverse
```

```python
import scipy.io as sio
import cv2
mdict = sio.loadmat("imgs_for_optical_flow.mat" )
img1=mdict['img1']
u=mdict['u']
v=mdict['v']


plt.subplot(1,3,1)
plt.imshow(img1, cmap="gray",interpolation=None)
plt.title("Original")

I_wrapped_1 = np.empty((80,80))
my_interpolation = bilinear_interpolation(lambda y,x: img1[x,y])
my_T_inverse = compute_T_inverse(v, u)

y_tilde, x_tilde = mgrid[0:80, 0:80]

T_minus_1_y, T_minus_1_x = my_T_inverse(y_tilde, x_tilde)
for j in range(80):
    for i in range(80):
        I_wrapped_1[i,j] = my_interpolation(T_minus_1_y[i,j],T_minus_1_x[i,j])
        
        
plt.subplot(1,3,2)
plt.imshow(I_wrapped_1, cmap="gray",interpolation=None)
plt.title("My Bilinear Interpolation", size=9)

img2 = np.empty((80,80))

yy,xx=mgrid[0:80,0:80]
U=xx-u
V=yy-v
img2 = cv2.remap(src =np.float32(img1), map1 = np.float32(U), map2 = np.float32(V), interpolation=cv2.INTER_LINEAR)
        
plt.subplot(1,3,3)
plt.imshow(img2, cmap="gray",interpolation=None)
plt.title("cv2.remap")

plt.savefig('computerExercise2.jpg',bbox_inches='tight')

plt.show()
```

```python
import cv2
import scipy.io as sio
mdict = sio.loadmat("imgs_for_optical_flow.mat" )


img1=mdict['img1']
img2=mdict['img2']
img3=mdict['img3']
img4=mdict['img4']
img5=mdict['img5']
img6=mdict['img6']

img = np.empty((6,80,80))
img[0] = img1
img[1] = img2
img[2] = img3
img[3] = img4
img[4] = img5
img[5] = img6

I_t = np.empty((5,80,80))
I_t[0] = img2-img1
I_t[1] = img3-img1
I_t[2] = img4-img1
I_t[3] = img5-img1
I_t[4] = img6-img1


I_x = np.empty((5,80,80))
I_x[0] = cv2.Sobel(src = img1, ddepth = -1 , dx=1, dy = 0)
I_x[1] = cv2.Sobel(src = img2, ddepth = -1 , dx=1, dy = 0)
I_x[2]= cv2.Sobel(src = img3, ddepth = -1 , dx=1, dy = 0)
I_x[3]= cv2.Sobel(src = img4, ddepth = -1 , dx=1, dy = 0)
I_x[4]= cv2.Sobel(src = img5, ddepth = -1 , dx=1, dy = 0)

I_y = np.empty((5,80,80))
I_y[0]= cv2.Sobel(src = img1, ddepth = -1 , dx=0, dy = 1)
I_y[1] = cv2.Sobel(src = img2, ddepth = -1 , dx=0, dy = 1)
I_y[2] = cv2.Sobel(src = img3, ddepth = -1 , dx=0, dy = 1)
I_y[3] = cv2.Sobel(src = img4, ddepth = -1 , dx=0, dy = 1)
I_y[4] = cv2.Sobel(src = img5, ddepth = -1 , dx=0, dy = 1)

def i_s(i,j):
    return 80*i+j
def col_i_j_u(i,j):
    return 2*i_s(i,j)
def col_i_j_v(i,j):
    return 2*i_s(i,j)+1
```

```python
import numpy as np
from scipy import sparse

A = np.empty((5,2*80*80,2*80*80))
b = np.empty((5,80*80*2))

lmd = 1
LT = 2*lmd
for pair in range(5):
    for i in range (80):
        for j in range (80):
            ###### CALCULATING b
            b[pair, col_i_j_u(i,j)] = -I_x[pair,i,j] * I_t[pair, i,j]
            b[pair, col_i_j_v(i,j)] = -I_y[pair, i,j] * I_t[pair, i,j]

            ###### CALCULATING A
            row_u=np.zeros(80*80*2)
            row_v=np.zeros(80*80*2)

            ### SELF

            row_u[col_i_j_u(i,j)] = I_x[pair, i,j] ** 2
            row_u[col_i_j_v(i,j)] = I_x[pair, i,j]*I_y[pair, i,j]

            row_v[col_i_j_u(i,j)] = I_x[pair, i,j]*I_y[pair, i,j]
            row_v[col_i_j_v(i,j)] = I_y[pair, i,j] ** 2

            ### NEIG

            if(i!=0):
                row_u[col_i_j_u(i-1,j)] = -LT
                row_u[col_i_j_u(i,j)] += LT

                row_v[col_i_j_v(i-1,j)] = -LT
                row_v[col_i_j_v(i,j)] += LT

            if(i!=79):
                row_u[col_i_j_u(i+1,j)] = -LT
                row_u[col_i_j_u(i,j)] += LT

                row_v[col_i_j_v(i+1,j)] = -LT
                row_v[col_i_j_v(i,j)] += LT

            if(j!=0):
                row_u[col_i_j_u(i,j-1)] = -LT
                row_u[col_i_j_u(i,j)] += LT

                row_v[col_i_j_v(i,j-1)] = -LT
                row_v[col_i_j_v(i,j)] += LT

            if(j!=79):
                row_u[col_i_j_u(i,j+1)] = -LT
                row_u[col_i_j_u(i,j)] += LT

                row_v[col_i_j_v(i,j+1)] = -LT
                row_v[col_i_j_v(i,j)] += LT



            A[pair, col_i_j_u(i,j)] = row_u
            A[pair, col_i_j_v(i,j)] = row_v
        
        
A_1_sparse = sparse.csr_matrix(A[0])
A_2_sparse = sparse.csr_matrix(A[1])
A_3_sparse = sparse.csr_matrix(A[2])
A_4_sparse = sparse.csr_matrix(A[3])
A_5_sparse = sparse.csr_matrix(A[4])

x = np.empty((5,80*80*2))
x[0]=sparse.linalg.lsqr(A_1_sparse, b[0])[0]
x[1]=sparse.linalg.lsqr(A_2_sparse, b[1])[0]
x[2]=sparse.linalg.lsqr(A_3_sparse, b[2])[0]
x[3]=sparse.linalg.lsqr(A_4_sparse, b[3])[0]
x[4]=sparse.linalg.lsqr(A_5_sparse, b[4])[0]


u = np.empty((5,80,80))
v = np.empty((5,80,80))

for pair in range(5):
    for i in range(80):
        for j in range(80):
            u[pair, i,j]=x[pair, col_i_j_u(i,j)]
            v[pair, i,j]=x[pair, col_i_j_v(i,j)]
```

```python
U = np.empty((5,80,80))
V = np.empty((5,80,80))
I_wrapped = np.empty((5,80,80))
yy,xx=np.mgrid[0:80,0:80]

for pair in range(5):
    U[pair] = xx-u[pair]
    V[pair] = yy-v[pair]
    
    
    I_wrapped[pair] = cv2.remap(src =np.float32(img1), map1 = np.float32(U[pair]),
                 map2 = np.float32(V[pair]), interpolation=cv2.INTER_LINEAR)
```

```python
cmap="gray"
def plot5(cmap, pair, img, u, v, I_wrapped):
    plt.subplot(1,5,1)
    plt.imshow(img[0],cmap=cmap)
    plt.title("img1")
    plt.axis("Off")

    plt.subplot(1,5,2)
    plt.imshow(u[pair],cmap=cmap)
    plt.title("u")
    plt.axis("Off")

    plt.subplot(1,5,3)
    plt.imshow(v[pair],cmap=cmap)
    plt.title("v")
    plt.axis("Off")

    plt.subplot(1,5,4)
    plt.imshow(img[pair+1],cmap=cmap)
    plt.title("img{}".format(pair+2))
    plt.axis("Off")

    plt.subplot(1,5,5)
    plt.imshow(I_wrapped[pair],cmap=cmap)
    plt.title("wrapped")
    plt.axis("Off")
    
    plt.savefig('computer_Exercise_5_pair_{}.jpg'.format(pair),bbox_inches='tight')
    
    
    plt.show()
```

```python
cmap = "gray"
for pair in range(5):
    plot5(cmap, pair, img, u, v, I_wrapped)
```
